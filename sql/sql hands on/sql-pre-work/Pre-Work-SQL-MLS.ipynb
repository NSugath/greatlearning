# <font color='blue'> Introduction </font>

In this session we are going to exploare and perform the various concepts related SQL databases using a database named as HR.

SQL file of this database is available on the Olympus for reference. In this session, we are going to execute DDL, DML, Joins and subqueries operations on the given database. 

An idea behind this session is to make our learner comfortable with the concepts such as: 
   
   * ER Diagram
   * Primary keys, foriegn keys
   * DDL and DML operations
   * Joins and Subqueries

To discuss the above mentioned topics we will be using MySQL workbench. Before beginning of this session please ensure that you have MySQL workbench is installed in your system and you are able to perform basic queries on the existing databases. 

Once the workbench is up and running, import the sql file to create the table and insert the pre-defined data points in the database.
